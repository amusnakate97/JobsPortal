### Outliers Detection and Handling

```python
cols=['age_first_funding_year', 'age_last_funding_year',
       'age_first_milestone_year', 'age_last_milestone_year', 'relationships',
       'funding_rounds', 'funding_total_usd', 'milestones','avg_participants']
```

```python
for i,col in enumerate(cols):
    plt.figure(i)
    sns.boxplot(df[col])
```

```python
def removeOutlier(col):
    sorted(col)
    quant1,quant2=col.quantile([0.25,0.75])
    IQR=quant2-quant1
    lowerRange=quant1-(1.5*IQR)
    upperRange=quant2+(1.5*IQR)
    return lowerRange,upperRange
```

```python
for col in cols:
    low,high=removeOutlier(df[col])
    df[col]=np.where(df[col]>high,high,df[col])
    df[col]=np.where(df[col]<low,low,df[col])
```

### Model Building

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
