### Recommendation Logic

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df.columns
```

```python
def combined_features(row):
    return row['Role Category']+" "+row['Location']+" "+row['Industry']+" "+row['Skill_1']+" "+row['Skill_2']+" "+row['Skill_3']

df["combined_features"] = df.apply(combined_features, axis =1)
```

```python
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
print("Count Matrix:", count_matrix.toarray())
```

```python
cosine_sim = cosine_similarity(count_matrix)
```

```python
job_searched = "python"
def get_index_from_title(title):
    return df[df["Skill_1"] == title].index.values[0]
job_index = get_index_from_title(job_searched)
```

```python
job_index
```

```python
similar_jobs = list(enumerate(cosine_sim[job_index]))
sorted_similar_jobs = sorted(similar_jobs, key=lambda x:x[1], reverse=True)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["Skill_1"].values[0]
i=0
for job in sorted_similar_jobs:
    print(get_title_from_index(job[0]))
    i=i+1
    if i>5:
        break
```
